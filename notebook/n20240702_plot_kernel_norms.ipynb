```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

norms = np.load('data\\multivariate_attempts\\btc_usd_hawkes_exp_kern_norm_matrix.npy')

event_type_times_map = pickle.load(open('data\\multivariate_attempts\\btc_usd_event_type_times_map.pkl', 'rb'))
def get_abbreviation(event_type: str) -> str:
    abbreviation = ''
    for substring in event_type.split('_'):
        abbreviation += substring[0]
    return abbreviation

event_type_abbreviations = [get_abbreviation(k) for k in event_type_times_map.keys()]
```

```python
df = pd.DataFrame(norms, columns=event_type_abbreviations, index=event_type_abbreviations)
order = ['ACONC', 'ALONC', 'AMONC', 'AMOC', 'ALOC', 'BLOC', 'BMOC', 'BMONC', 'BLONC', 'BCONC']
df = df[order]
df = df.reindex(order)
df
```

```python
sns.heatmap(df, cmap="Blues")

plt.savefig('images\\multivariate_attempts\\btc_usd_hawkes_exp_kern_norm_matrix.svg')
plt.savefig('images\\multivariate_attempts\\btc_usd_hawkes_exp_kern_norm_matrix.pdf')

plt.show()
```
